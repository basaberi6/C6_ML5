```python
# Clean date columns
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Create 'Year' column

df['Year'] = df['Order Date'].dt.to_period('Y').astype(str)
```

```python
# Getting top 10 most profitable orders
Top_10=df.sort_values('Profit',ascending=False).reset_index().head(10)
print(Top_10)
```

```python
# Getting orders with Sales > $1000
sales_amt_high=df['Order ID'][df['Sales']>1000.0]
print(sales_amt_high)
```

```python
#Calculating additional business metrics
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
total_orders = df['Order ID'].nunique()
avg_discount = df['Discount'].mean()

print(f" Total Sales: {total_sales:,.2f}")
print(f" Total Profit: {total_profit:,.2f}")
print(f" Total Orders: {total_orders}")
print(f" Avg. Discount: {avg_discount:.2%}")
```

```python

monthly_sales = df.groupby('Year')['Sales'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(monthly_sales['Year'], monthly_sales['Sales'], marker='o')
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group sales by region
region_sales = df.groupby('Region')['Sales'].sum().reset_index()

# Plot pie chart
plt.figure(figsize=(5, 5))
plt.pie(region_sales['Sales'], labels=region_sales['Region'], autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution by Region')
plt.axis('equal')  # Equal aspect ratio ensures the pie is a circle
plt.tight_layout()
plt.show()
```

```python
# Group sales by category
category_sales = df.groupby('Category')['Sales'].sum().reset_index()

# Sort values (optional, for nicer chart)
category_sales = category_sales.sort_values(by='Sales', ascending=False)

# Plot
plt.figure(figsize=(6, 4))
plt.bar(category_sales['Category'], category_sales['Sales'], color='skyblue')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#Sales by state
plt.figure(figsize=(14, 6))
plt.bar(df['State'], df['Sales'], color='teal')
plt.title('Total Sales by State')
plt.xlabel('State')
plt.ylabel('Sales')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
#Check for losses
loss_data = df[df["Profit"] < 0]
print("Sample of Loss Data:")
print(loss_data.head())
```

```python
#Sales and Profit by Category
category_summary = df.groupby('Category')[['Sales', 'Profit']].sum().reset_index()
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = category_summary['Category']
sales = category_summary['Sales']
profit = category_summary['Profit']

# Bar positions
x = np.arange(len(categories))  # the label locations
width = 0.35  # the width of the bars

# Plot
plt.figure(figsize=(6, 4))
plt.bar(x - width/2, sales, width, label='Sales', color='skyblue')
plt.bar(x + width/2, profit, width, label='Profit', color='orange')

# Labels & Legend
plt.xlabel('Category')
plt.ylabel('Amount')
plt.title('Sales and Profit by Category')
plt.xticks(x, categories)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
product_profit = df.groupby('Product Name')['Profit'].sum().reset_index()
product_profit = product_profit.sort_values(by='Profit', ascending=False)
print("Top 10 Profitable Products:")
print(product_profit.head(10))

print("\nBottom 10 (Most Unprofitable) Products:")
print(product_profit.tail(10))
```
